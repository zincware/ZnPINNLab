# Automating the process
We are now able to create training dictionaries by hand. But what if we want to create several of them? The hard way is to create each of them by hand. The smart way on the other side is to automate the process. In order to do so we split the process in two steps:

### Task 4
1. **Build a `create_training_data` function:**
   - Combine the steps of Task 2 and 3. More informations on the function are given in the documentation string.
   - Generate and print a `train_dict` by utilizing the `create_training_data` function.
   <br><br>
2. **Generate a `data_dict` that contains all `train_dicts`:**
   - Create a List called `looplist` with 3 tuples of your choice associated with (`t_min`, `t_max`).
   - Set up an empty dictionary called `data_dict`
   - Build a For-loop that iterates over the tuples (`t_min`, `t_max`) in the looplist. Use the enumerate Method to get a counter
   - Inside the loop, 
      - Define a variable called `train_dict` and utilize it to store a dictionary. Generate this dictionary by calling the `create_training_data` function based on `t_min` and `t_max`.
      - Define a variable called key, that stores dynamically generated keys. Consider using a formatted string (e.g., `"data_0"`, `"data_1"`, `"data_2"`) to create unique keys for each iteration.
      - Store the generated data and its associated key in the `data_dict`. 
   - Outside the loop, print or display the `data_dict` to view the stored data.

````python
def create_training_data(t_min, t_max):
    """
    Creates training data for a machine learning model.

    Parameters
    ----------
    t_min : int
        The minimum value for the input data.
    t_max : int
        The maximum value for the input data.

    Returns
    -------
    dict
        A dictionary containing the input and target data.
        The dictionary has two entries:
        - "input": a tensor containing the input data
        - "target": a tensor containing the target data
    """
    steps = t_max - t_min + 1
    
    data_input = torch.linspace(t_min, t_max, steps)
    data_target = torch.linspace(t_min*2, t_max*2, steps)

    return {"input": data_input, "target": data_target}

train_dict = create_training_data(t_min=2, t_max=3)
print("train_dict: ", train_dict)
````

````python
looplist = [(1, 10), (5, 10), (2, 12)]
data_dict = {}
for i, (t_min, t_max) in enumerate(looplist):
    train_dict = create_training_data(t_min, t_max)
    key = f"data_{i}"  # Use a formatted string to create a unique key
    data_dict[key] = train_dict
data_dict
````
