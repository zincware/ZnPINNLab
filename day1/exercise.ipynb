## Understanding Dictionaries in Python

A **dictionary** in Python is a versatile data structure that allows you to store and manage key-value pairs.<br> Each key is associated with a specific value, making it easy to organize and retrieve data efficiently.<br> Dictionaries are enclosed in curly braces `{}` and consist of key-value pairs separated by colons `:`.

### Components of a Dictionary

1. **Keys**: Keys are unique identifiers that help access specific values in a dictionary. Keys are typically strings but can also be numbers or other immutable data types.

2. **Values**: Values are the associated data or information corresponding to each key. Values can be of any data type, including numbers, strings, lists, or even other dictionaries.

## Creating a Dictionary

Let's illustrate how to create a dictionary with a simple example:


````python
# Define a dictionary with key-value pairs
example_dict = {
    "name": "John",
    "age": 30,
    "city": "New York"
}
````

One way of accessing a key is to store its value in a variable:

````python
name = example_dict["name"]
age = example_dict["age"]

print("The value associated with the key \"name\" is:", name)
print("The value associated with the key \"name\" is:", age)
````
